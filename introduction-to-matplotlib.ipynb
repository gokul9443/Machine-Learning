### Matplotlib example workflow (with code)

```python
# 0. import matplotlib and get in ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup Plot
fig, ax = plt.subplots(figsize = (10, 10)) # figsize = (10, 5) - width and height

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title = "Simple Plot", xlabel = "x-axis", ylabel = "y-axis")

#5. Save and show (Save the whole figure)
fig.savefig("matplotlib_images/sample-plot.png")
```

### Two options for subplots

```python
# Subplots option 1 <------- Recommended
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols = 2, figsize = (10,5))

np.random.seed(7)
# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax4.hist(x);
```

```python
# Subplots option 2
fig, ax = plt.subplots(nrows = 2, ncols = 2, figsize = (10,5))

np.random.seed(7)
# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(),nut_butter_prices.values());
ax[1, 1].hist(x);
```

### Which one should you use? (pyplot vs matplotlib OO method?)
* When plotting something very quickly, use pyplot method
* When pltiing something more advanced, use OO method

```python
over_50 = heart_disease[heart_disease['age']>50]
```

```python
over_50
```

#### Pyplot method

```python
over_50.plot(kind = 'scatter', x = "age", y = "chol", c = "target");
```

#### OO Method mixed with pyplot

```python
fix, ax_1 = plt.subplots(figsize = (10,6))

over_50.plot(kind = 'scatter', x = "age", y = "chol", c = "target", ax = ax_1);

ax_1.set_xlim([45,100]);
```

#### OO Method 

```python
## OO method from scratch
fig, ax_2 = plt.subplots(figsize = (10,6), sharex = True)

# Plot the data
scatter = ax_2.scatter(x = over_50["age"], y = over_50["chol"], c = over_50['target']); # scatter obj used later to customize the plot

# customize the plot
ax_2.set(title = "Heart Disease and Cholestrol Levels", xlabel = "Age", ylabel = "Cholestrol");

# Add a legend
ax_2.legend(*scatter.legend_elements(), title = "Target");

# Add a horizontal line
ax_2.axhline(over_50["chol"].mean(), ls = '--');
```

```python
# Subplot of chol, age, thalach
fig, ((ax1, ax2)) = plt.subplots(nrows = 2, ncols = 1, figsize = (10,10), sharex = True) 

# Add data to ax1
scatter = ax1.scatter(x = over_50["age"], y = over_50["chol"], c = over_50['target']);

# Customize ax1
ax1.set(title = "Heart Disease and Cholestrol Levels", ylabel = "Cholestrol");

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title = "Target");

# Add a horizontal line to ax1
ax1.axhline(over_50["chol"].mean(), ls = '--');

# Add data to ax2
scatter2 = ax2.scatter(x = over_50["age"], y = over_50["thalach"], c = over_50['target']); # scatter obj used later to customize the plot

# Customize ax2
ax2.set(title = "Heart Disease and Max Heart Rate", xlabel = "Age", ylabel = "Max Heart Rate");

# Add a legend to ax1
ax2.legend(*scatter2.legend_elements(), title = "Target");

# Add a horizontal line to ax1
ax2.axhline(over_50["thalach"].mean(), ls = '--');

# Add title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```

```python
# Set the style
plt.style.use('seaborn-whitegrid')

## OO method from scratch
fig, ax_2 = plt.subplots(figsize = (10,6), sharex = True)

# Plot the data
scatter = ax_2.scatter(x = over_50["age"], y = over_50["chol"], c = over_50['target'], cmap = "winter"); # cmap - color
# scatter obj used later to customize the plot

# customize the plot
ax_2.set(title = "Heart Disease and Cholestrol Levels", xlabel = "Age", ylabel = "Cholestrol");

# Add a legend
ax_2.legend(*scatter.legend_elements(), title = "Target");

# Add a horizontal line
ax_2.axhline(over_50["chol"].mean(), ls = '--');
```

```python
# customize x and y axis limitations and make the plot neat and clean


# Subplot of chol, age, thalach
fig, ((ax1, ax2)) = plt.subplots(nrows = 2, ncols = 1, figsize = (10,10), sharex = True) 

# Add data to ax1
scatter = ax1.scatter(x = over_50["age"], y = over_50["chol"], c = over_50['target'], cmap = "winter");

# Customize ax1
ax1.set(title = "Heart Disease and Cholestrol Levels", ylabel = "Cholestrol");

# change the axis limits
ax1.set_xlim([50,80])
ax1.set_ylim([0, 600])

# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title = "Target");

# Add a horizontal line to ax1
ax1.axhline(over_50["chol"].mean(), ls = '--');

# Add data to ax2
scatter2 = ax2.scatter(x = over_50["age"], y = over_50["thalach"], c = over_50['target'], cmap = "winter"); # scatter obj used later to customize the plot

# Customize ax2
ax2.set(title = "Heart Disease and Max Heart Rate", xlabel = "Age", ylabel = "Max Heart Rate");

# change the axis limits
ax2.set_xlim([50,80])
ax2.set_ylim([20,300])

# Add a legend to ax1
ax2.legend(*scatter2.legend_elements(), title = "Target");

# Add a horizontal line to ax1
ax2.axhline(over_50["thalach"].mean(), ls = '--');

# Add title to the figure
fig.suptitle("Heart Disease Analysis", fontsize = 16, fontweight = "bold");
```
